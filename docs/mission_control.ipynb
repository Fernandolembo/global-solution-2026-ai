```python
import random

def gerar_dados_missao():
    """Simula dados da telemetria espacial."""
    return {
        "temperatura": random.randint(20, 120),
        "energia": random.randint(5, 100),
        "comunicacao": random.choice(["Estável", "Instável", "Crítica", "Sinal Fraco"]),
        "status_modulos": "Operacional"
    }

def analisar_com_ia(dados):
    """Envia os dados para o Llama 3.2 analisar."""
    prompt_sistema = """
    Você é o computador central da Missão Mission Control AI.
    Analise os dados de telemetria e forneça um relatório curto (máx 3 linhas).
    Se houver perigo, use um tom urgente. Se estiver tudo bem, seja formal.
    """

    conteudo_usuario = f"""
    STATUS ATUAL:
    - Temperatura: {dados['temperatura']}°C
    - Energia: {dados['energia']}%
    - Comunicação: {dados['comunicacao']}
    - Módulos: {dados['status_modulos']}
    Avalie a situação e dê recomendações.
    """

    response = ollama.chat(
        model="llama3.2:1b",
        messages=[
            {"role": "system", "content": prompt_sistema},
            {"role": "user", "content": conteudo_usuario}
        ]
    )
    return response["message"]["content"]

def exibir_dashboard(dados, analise):
    """Exibe os dados de forma organizada (conforme critério de avaliação)."""
    print("\n" + "="*50)
    print("        🚀 MISSION CONTROL AI - FIAP 2026")
    print("="*50)

    alerta = " STATUS: NORMAL"
    if dados['temperatura'] > 90:
        alerta = " ALERTA: SUPERAQUECIMENTO DETECTADO!"
    elif dados['energia'] < 20:
        alerta = " CRÍTICO: MODO DE ECONOMIA ATIVADO!"
    elif dados['comunicacao'] == "Crítica":
        alerta = " ERRO: PERDA DE SINAL IMINENTE!"

    print(f" {alerta}")
    print("-" * 50)
    print(f" [TEMP]     : {dados['temperatura']}°C")
    print(f" [ENERGIA]  : {dados['energia']}%")
    print(f" [COMMS]    : {dados['comunicacao']}")
    print(f" [MÓDULOS]  : {dados['status_modulos']}")
    print("-" * 50)
    print("  ANÁLISE DA INTELIGÊNCIA ARTIFICIAL:")
    print(f" {analise}")
    print("="*50 + "\n")
```

```python
import time

N_CICLOS = 5

print(" INICIANDO MONITORAMENTO DA MISSÃO ESPACIAL...")
print("Aguardando conexão com os módulos...\n")

for i in range(1, N_CICLOS + 1):
    print(f"📡 Lendo dados do Ciclo {i}/{N_CICLOS}...")

    # 1. Gerar dados aleatórios
    dados_atuais = gerar_dados_missao()

    # 2. Obter análise da Inteligência Artificial
    try:
        analise_ia = analisar_com_ia(dados_atuais)
    except Exception as e:
        analise_ia = f"Erro ao conectar com a IA: {e}"

    # 3. Exibir o Dashboard formatado
    exibir_dashboard(dados_atuais, analise_ia)

    # 4. Pausa de 3 segundos entre as leituras para dar efeito de "tempo real"
    if i < N_CICLOS:
        print("Aguardando próxima telemetria...\n")
        time.sleep(3)

print("✅ Monitoramento finalizado. Logs salvos para o Centro de Comando.")
```
